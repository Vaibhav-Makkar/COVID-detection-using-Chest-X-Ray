```python
import tensorflow as tf
from keras.preprocessing import image
model = tf.keras.models.load_model(r'C:\Users\Vaibhav Makkar\Desktop\COVID 19 Chest XRay\model.h5')
from keras.applications.vgg16 import preprocess_input
```

```python
img=image.load_img(r'C:\Users\Vaibhav Makkar\Desktop\COVID 19 Chest XRay\Data\Covid\16660_1_1.jpg',target_size=(224,224,3))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# Convert the image to array to be given to model
import numpy as np
X=image.img_to_array(img)
X=np.expand_dims(X,axis=0)
```

```python
X
```

```python
img_data = preprocess_input(X)
```

```python
img_data
```

```python
classes = model.predict(img_data)
New_pred = np.argmax(classes,axis=1)
if(New_pred==[1]):
    print('Prediction Normal')
else:
    print('Prediction Covid')
```
